```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
       'the glass of juice',
       'the cup of tea',
       'i am a good boy',
       'i am an AI Engineer',
       'understand the meaning of words',
       'your videos are good']
```

```python
sent
```

```python
# Vocablary_size
voc_size = 10000
```

## One_Hot Representation

```python
onehot_repr = [one_hot(words, voc_size) for words in sent]
onehot_repr
```

## Word Embadding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

1. length of all sentences most be same for pad_sequences
2. `pre` we have some sentences that have one extra word or less for that we use use `pre` pad_sequence
3. sentance_length is 8.. we can set any number accurding to our sentence
4. our sentence lenght is less than 8 but we still pass 8
5. padding=`pre` will set 0 for extra length.. you can see in the below cell

```python
sent_length = 8
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
embedded_docs
```

6. Now try padding='`post`'

```python

embedded_docss = pad_sequences(onehot_repr, padding='post', maxlen=sent_length)
embedded_docss
```

* Now pass this to the embedding layer
* for that we need dimension.. let's declare

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length=sent_length))
model.compile('Adam', 'mse') #mse is mean squared error
```

`mse` is mean squared error

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
